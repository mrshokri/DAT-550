```python
# Load the test dataset
test_data = pd.read_csv("C:/Users/pc/Desktop/bow_test.csv")
test_data 
```

```python
# Tokenize the abstracts into words or tokens for train and test sets
train_abstracts = train_data['abstract'].tolist()
test_abstracts = test_data['abstract'].tolist()
```

```python
# Convert the abstracts into bags-of-words representations using CountVectorizer for train and test sets
count_vectorizer_train = CountVectorizer()
X_train_count = count_vectorizer_train.fit_transform(train_abstracts)

count_vectorizer_test = CountVectorizer(vocabulary=count_vectorizer_train.vocabulary_)
X_test_count = count_vectorizer_test.transform(test_abstracts)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit the vectorizer on the training data and transform it
X_train_tfidf = tfidf_vectorizer.fit_transform(train_abstracts)

# Transform the test data using the fitted vectorizer
X_test_tfidf = tfidf_vectorizer.transform(test_abstracts)
```

```python
# Get the corresponding labels for train and test sets
y_train = train_data['label']
y_test = test_data['label']
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define FFNN model
class FFNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(FFNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# Initialize parameters
input_size = X_train_tfidf.shape[1]  # Size of input features (TF-IDF vectors)
hidden_size = 128  # Size of the hidden layer
num_classes = len(train_data['label'].unique())  # Number of output classes

# Create an instance of the FFNN model
model = FFNN(input_size, hidden_size, num_classes)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```
